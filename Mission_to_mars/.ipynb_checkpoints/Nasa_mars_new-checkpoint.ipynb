```python
#import dependencies
from bs4 import BeautifulSoup
from splinter import Browser
from pprint import pprint
import pandas as pd
import pymongo
import requests
```

```python
#define browser
executable_path = {'executable_path':'C:\\chromedriver_win32\\chromedriver.exe' }
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://mars.nasa.gov/news/'
browser.visit(url)
```

```python
#iterate through all pages
#For this code  igot till the link is displayed

for page in range(1,10):
    #HTML Object
    html = browser.html
    #parse HTML with Beautiful soup
    soup = BeautifulSoup(html, 'html.parser')
    
    content_titles = soup.find_all('div', class_='content_title')[1:]
    
    for title in content_titles:
#        print(title.prettify())
       link =title.find('a')
#        print(link)
       news_url=link['href']
#        news_url= link.find('a')['href']
       print(news_url)
      
    
   
```
